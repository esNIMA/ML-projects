```python
import pandas as pd 
import numpy as np 
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import TomekLinks 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("creditcard.csv")
df
```

```python
df["Class"].value_counts()
```

### Removing duplicates 

```python
df = df.drop_duplicates()
```

```python
df.drop("Time" , axis=1 , inplace=True)
```

```python
# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Fit the encoder on the "Class" column and transform it
df['Class'] = encoder.fit_transform(df['Class'])
```

### Evaluating acc with all 0 array 

```python
accuracy_score(df["Class"], np.zeros_like(df["Class"]))
```

```python
report = classification_report(df["Class"], np.zeros_like(df["Class"]))
print("Classification Report:\n", report)
```

## Train test split 

```python
# Divide data into training and validation subsets
X_train, X_valid, y_train, y_valid = train_test_split(df.drop(["Class"], axis=1), df["Class"], train_size=0.8, test_size=0.2,random_state=0)
```

### Under-sampling

```python
rus = TomekLinks()

X_train, y_train = rus.fit_resample(X_train, y_train)
```

```python
model = SVC()
```

```python
model.fit(X_train , y_train)
```

```python
y_preds = model.predict(X_valid)
```

```python
accuracy_score(y_valid , y_preds)
```

```python
report = classification_report(y_valid, y_preds)
print("Classification Report:\n", report)
```
